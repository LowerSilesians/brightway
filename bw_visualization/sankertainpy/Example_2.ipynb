### Generate distribution plot with marked quantiles from sankey diagram type 1:

```python
data=result['scores'][0]
```

```python
fig2 = ff.create_distplot([data], ['ta'], bin_size=0.0007, show_hist=True, show_rug=False)
```

```python
cmap = matplotlib.cm.get_cmap('RdYlGn')
cmap_basic = matplotlib.cm.get_cmap('Blues')
quantiles= np.arange(0.125,0.925,0.05) 
```

```python
old_quant=0
last_x=[]
last_y=[]
for qu in quantiles:
    xl = np.quantile(data, old_quant)
    xr = np.quantile(data, qu)
    x1=last_x
    y1=last_y

    for i, xc in enumerate(fig2.data[0].x):
        if xc >xl and xc <xr:
            x1.append(xc)
            y1.append(fig2.data[0].y[i]) 
        
    if old_quant==0:
        fig2.add_scatter(x=x1, y=y1,fill='tozeroy', mode='none' , fillcolor=str('rgba'+str(cmap_basic(0.5,1.0))))
    else:
        fig2.add_scatter(x=x1, y=y1,fill='tozeroy', mode='none' , fillcolor=str('rgba'+str(cmap(1-qu,0.9))))
    last_x=[x1[len(x1)-1]]
    last_y=[y1[len(x1)-1]]
    old_quant = qu
```

```python
fig2.layout.paper_bgcolor='rgba(0,0,0,0)'
fig2.update_yaxes(visible=True)
fig2.update_layout(showlegend=False)
```
